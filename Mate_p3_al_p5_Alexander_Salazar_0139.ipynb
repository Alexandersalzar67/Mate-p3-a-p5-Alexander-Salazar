#### Problema 3: El número misterioso
#### Planteamiento:
* Llamemos $x$ al número que buscamos.
* "Resta siete": $x - 7$
* "Da como resultado 18": $= 18$
#### Ecuación:
$$x - 7 = 18$$
#### Despeje:
1. Sumamos 7 en ambos lados para dejar a la $x$ sola.
2. $x = 18 + 7$
3. Resultado: $x = 25$

```python
# Problema 3: x - 7 = 18
resultado_final = 18
resta = 7

# Despejamos sumando el valor restado al resultado
numero = resultado_final + resta

print(f"El número buscado en el problema 3 es: {numero}")
```

#### Problema 4: El triple de un número
#### Planteamiento:
* "El triple de un número": $3x$
* "Más 4": $+ 4$
* "Igual a 31": $= 31$Ecuación:$$3x + 4 = 31$$
#### Despeje:
1. Restamos 4 en ambos lados: $3x = 31 - 4$
2. Simplificamos: $3x = 27$
3. Dividimos entre 3: $x = 27 / 3$
4. Resultado: $x = 9$

```python
# Problema 4: 3x + 4 = 31
total = 31
suma = 4
multiplo = 3

# Primero restamos el valor que suma
paso_1 = total - suma
# Luego dividimos por el multiplicador (triple)
x = paso_1 / multiplo

print(f"El número buscado en el problema 4 es: {int(x)}")
```

Problema 5: Dimensiones del terreno

Planteamiento:

* Ancho: $a$
* Largo (5 metros más que el ancho): $a + 5$
* Fórmula del perímetro: $P = 2(largo) + 2(ancho)$
* Perímetro total: $50$
#### Ecuación:
$$2(a + 5) + 2(a) = 50$$
#### Despeje:
1. Distribuimos el 2: $2a + 10 + 2a = 50$
2. Agrupamos términos semejantes: $4a + 10 = 50$
3. Restamos 10: $4a = 40$Dividimos entre 4: $a = 10$
4. Resultado: El ancho mide 10 m y el largo mide 15 m ($10 + 5$).

```python
# Problema 5: 2(a + 5) + 2a = 50 -> 4a + 10 = 50
perimetro_total = 50
diferencia_largo = 5

# El perímetro es 2*ancho + 2*(ancho + 5) = 4*ancho + 10
# Despejamos el ancho (a)
ancho = (perimetro_total - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"Dimensiones del terreno:")
print(f"Ancho: {ancho} m")
print(f"Largo: {largo} m")
```
